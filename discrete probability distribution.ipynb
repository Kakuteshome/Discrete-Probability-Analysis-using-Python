```python
df = pd.DataFrame({'X':[0,1,2,3],'P(x)':[0.2,0.4,0.3,0.1]})
e_x = sum(df['X']*df['P(x)'])
print('E(X):',e_x)
df['(X - E(X))**2'] = (df['X'] - e_x)**2
print(df)
print('sigma**2:',sum(df['(X - E(X))**2']*df['P(x)']))
print('sigma:',np.sqrt(sum(df['(X - E(X))**2']*df['P(x)'])))
```

```python
# Binom distribution

# Define parameters
n = 4  # Number of trials
p = 0.5  # Probability of success
k = 2  # Number of successes

# Calculate binomial probability
probability = binom.pmf(k, n, p)
print("Binomial probability:", probability)
```

```python

# Define parameters
n = 1500  # Number of trials
p = 0.156  # Probability of success

# Initialize lists to store results
successes = []
probabilities = []

# Calculate binomial probability for different numbers of successes
for k in range(n + 1):
    success_prob = binom.pmf(k, n, p)
    successes.append(k)
    probabilities.append(success_prob)

# Create DataFrame
df = pd.DataFrame({"Successes": successes, "Probability": probabilities})
print(df)
# Plot the binomial distribution
plt.bar(df["Successes"], df["Probability"])
plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.title(f"Binomial Distribution (n={n}, p={p})")
plt.show()
```

```python
# Parameters
n = 10  # Number of trials
p = 0.6  # Probability of success

# Calculate the first few probabilities
k_values = np.arange(0, n+1)  # Values of k from 0 to 10
probabilities = [binom.pmf(i, n, p) for i in k_values]

mean = n * p
variance = n * p * (1 - p)

print("\nMean:", mean)
print("Variance:", variance)

df = pd.DataFrame({'k':k_values,'P(k)':probabilities})
df['cumsum'] = 1-df['P(k)'].cumsum()
print(df)
# Generate the range of values for the x-axis (number of successes)
x = np.arange(0, n+1)

# Calculate the CDF for each value
cdf = binom.cdf(x, n, p)

# Plot the more than CDF
plt.step(x, df['cumsum'], where='post', color='b')
plt.xlabel('Number of Successes (k)')
plt.ylabel('Cumulative Probability')
plt.title('CDF of a Binomial Distribution (n={}, p={})'.format(n, p))
plt.grid(True)
plt.show()

```

```python
# Define parameters
N = 50  # Total number of items in the population
K = 40  # Total number of successes in the population
n = 5  # Number of items drawn in each trial

# Generate possible values for the number of successes in the sample
x = np.arange(0,n+1)

# Calculate the probability mass function (PMF)
pmf = hypergeom.pmf(x, N, K, n)

# Print the PMF
print("x\tp(x)")
print("-" * 15)
for i in range(len(x)):
    print(f"{x[i]}\t{pmf[i]:.5f}")

# Plot the PMF
plt.bar(x, pmf, align="center", alpha=0.7)
plt.xlabel("Number of Successes in Sample")
plt.ylabel("Probability")
plt.title("Hypergeometric Distribution PMF")
plt.show()
```

```python
# Parameters
average_rate = 10  # Average number of events per interval

# Generate a Poisson distribution
x = np.arange(0, 21)  # Number of events
pmf = poisson.pmf(x, average_rate)  # Probability mass function

# Plot the Poisson distribution
plt.bar(x, pmf, color='skyblue')
plt.xlabel('Number of Events')
plt.ylabel('Probability')
plt.title('Poisson Distribution')
plt.grid(True)
plt.show()
```

```python

# Set the lambda parameter for the Poisson distribution
lambda_parameter = 0.7

# Generate random samples from the Poisson distribution
samples = np.random.poisson(lambda_parameter, 1000)

# Calculate the probability mass function (PMF) for the Poisson distribution
x_values = np.arange(0, 10)
pmf_values = poisson.pmf(x_values, lambda_parameter)
print(pd.DataFrame({'X':x_values,'P(X)':pmf_values}))

# Plot the histogram of the generated samples
plt.hist(samples, bins=range(11), density=True, alpha=0.5, label='Sampled Data',edgecolor = 'black')

# Plot the PMF of the Poisson distribution
plt.plot(x_values, pmf_values, marker='o', linestyle='-', label='Poisson PMF', color='red')

# Add labels and legend
plt.xlabel('Number of Events')
plt.ylabel('Probability')
plt.title('Poisson Distribution Analysis')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```
